```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# Use a built-in style that's available everywhere
plt.style.use('classic')

# Create sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(-x/5)

# Create a figure with subplots using GridSpec
fig = plt.figure(figsize=(12, 8))
gs = GridSpec(2, 2, figure=fig)

# 1. Line plot with multiple lines and legend
ax1 = fig.add_subplot(gs[0, 0])
ax1.plot(x, y1, label='sin(x)', color='blue', linewidth=2)
ax1.plot(x, y2, label='cos(x)', color='red', linewidth=2)
ax1.set_title('Trigonometric Functions', pad=10)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()
ax1.grid(True, alpha=0.3)

# 2. Scatter plot with color mapping
ax2 = fig.add_subplot(gs[0, 1])
scatter = ax2.scatter(x, y3, c=x, cmap='viridis', s=50)
ax2.set_title('Exponential Decay', pad=10)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
plt.colorbar(scatter, ax=ax2, label='x value')
ax2.grid(True, alpha=0.3)

# 3. Bar plot
ax3 = fig.add_subplot(gs[1, 0])
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]
bars = ax3.bar(categories, values, color='lightblue', edgecolor='black')
ax3.set_title('Bar Chart Example', pad=10)
ax3.set_xlabel('Categories')
ax3.set_ylabel('Values')
# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    ax3.text(bar.get_x() + bar.get_width()/2., height,
             f'{height}', ha='center', va='bottom')
ax3.grid(True, axis='y', alpha=0.3)

# 4. Histogram with KDE
ax4 = fig.add_subplot(gs[1, 1])
data = np.random.normal(0, 1, 1000)
ax4.hist(data, bins=30, density=True, alpha=0.7, color='lightblue', 
         edgecolor='black')
# Add KDE
from scipy.stats import gaussian_kde
kde = gaussian_kde(data)
x_kde = np.linspace(min(data), max(data), 100)
ax4.plot(x_kde, kde(x_kde), 'r-', lw=2, label='KDE')
ax4.set_title('Histogram with KDE', pad=10)
ax4.set_xlabel('Value')
ax4.set_ylabel('Density')
ax4.legend()
ax4.grid(True, alpha=0.3)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show plot
plt.show()

# Optional: Save the figure
# plt.savefig('matplotlib_examples.png', dpi=300, bbox_inches='tight')
```
